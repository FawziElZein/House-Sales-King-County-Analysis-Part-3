### bedrooms vs sqft space analysis

The purpose of the below study is to proof by data that bedrooms is a categorical variable by trying to get the relationship between this variable and the house space.

```python
df[['id','bedrooms','sqft_living','sqft_lot','sqft_above','sqft_basement','yr_built']].sort_values('bedrooms',ascending =False).head(15)
```

```python
# the range of sqft_living of houses with bedrooms>7
[df[df['bedrooms']>7]['sqft_living'].min(),df[df['bedrooms']>7]['sqft_living'].max()]
```

```python
# the range of sqft_living of all houses
[df['sqft_living'].min(),df['sqft_living'].max()]
```

```python
df.groupby('bedrooms').agg({'sqft_living':['max','min']}).sort_values('bedrooms',ascending =False)
```
